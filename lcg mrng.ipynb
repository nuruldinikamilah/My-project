```python
def linear_congruential_generator(a, c, mod, Z, num_randoms):
    random_numbers = []

    for _ in range(num_randoms):
        Z = (a * Z + c) % mod
        random_numbers.append(Z)

    return random_numbers

a = 2
c = 5
mod = 100
Z = 10121216
num_randoms = 10

random_numbers = linear_congruential_generator(a, c, mod, Z, num_randoms)

print("Bilangan acak:")
for num in random_numbers:
    print(num)
```
